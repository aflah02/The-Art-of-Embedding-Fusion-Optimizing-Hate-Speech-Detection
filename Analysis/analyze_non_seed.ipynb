```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
result_dir = '../Models/NonSeed_Results/'
```

```python
files = os.listdir(result_dir)
```

```python
def open_json(file):
    with open(file, encoding='utf-8') as f:
        data = json.load(f)
    return data
```

```python
data_dynahate = {}
data_olid = {}
data_latenthatred = {}
for file in files:
    if file == 'README.md':
        continue
    if 'dynahate' in file:
        data_dynahate[file] = open_json(result_dir+file)
    elif 'olid' in file:
        data_olid[file] = open_json(result_dir+file)
    elif 'latenthatred' in file:
        data_latenthatred[file] = open_json(result_dir+file)
```

```python
df_dynahate = pd.DataFrame(data_dynahate).T
df_dynahate = df_dynahate[['Accuracy Test', 'Macro F1 Test']]
df_dynahate = df_dynahate.sort_values(by='Accuracy Test', ascending=False)
df_dynahate.index = df_dynahate.index.str.replace('_dynahate', '')

df_olid = pd.DataFrame(data_olid).T
df_olid = df_olid[['Accuracy Test', 'Macro F1 Test']]
df_olid = df_olid.sort_values(by='Accuracy Test', ascending=False)
df_olid.index = df_olid.index.str.replace('_olid', '')

df_latenthatred = pd.DataFrame(data_latenthatred).T
df_latenthatred = df_latenthatred[['Accuracy Test', 'Macro F1 Test']]
df_latenthatred = df_latenthatred.sort_values(by='Accuracy Test', ascending=False)
df_latenthatred.index = df_latenthatred.index.str.replace('_latenthatred', '')
```

```python
df_dynahate.plot(kind='bar', figsize=(15, 5))
```

```python
df_olid.plot(kind='bar', figsize=(15, 5))
```

```python
# plot
plt.figure(figsize=(10, 5))
plt.title('DynaHate')
plt.xlabel('Model')
# Plot Accuracy and Macro F1 as side by side bars
plt.bar(df_dynahate.index, df_dynahate['Accuracy Test'], label='Accuracy')
plt.bar(df_dynahate.index, df_dynahate['Macro F1 Test'], label='Macro F1')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# plot
plt.figure(figsize=(10, 5))
plt.title('OLID')
plt.xlabel('Model')
# Plot Accuracy and Macro F1 as side by side bars
plt.bar(df_olid.index, df_olid['Accuracy Test'], label='Accuracy')
plt.bar(df_olid.index, df_olid['Macro F1 Test'], label='Macro F1')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# plot
plt.figure(figsize=(10, 5))
plt.title('LatentHatred')
plt.xlabel('Model')
# Plot Accuracy and Macro F1 as side by side bars
plt.bar(df_latenthatred.index, df_latenthatred['Accuracy Test'], label='Accuracy')
plt.bar(df_latenthatred.index, df_latenthatred['Macro F1 Test'], label='Macro F1')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
